# Feature Extraction

Mel Frequency Cepstral Coefficents (MFCCs) are a feature widely used in automatic speech and speaker recognition.

The Mel-scale aims to mimic the non-linear human ear perception of sound, by being more discriminative at lower frequencies and less discriminative at higher frequencies
$$m = 2595log_{10}(1 + \frac{f}{700})$$

```python
%matplotlib inline
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# https://cdn.echo.msk.ru/snd/2018-08-26-beseda-1616.mp3
y, sr = librosa.load('data/2018-08-26-beseda-1616.mp3', sr=None, duration=10)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(y)
plt.title('Signal amplitude')
p = plt.gca()
p.yaxis.tick_right()
p.yaxis.set_label_position("right")
p.set_xlim(0, sr*10);
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr)
mfcc = librosa.feature.mfcc(S=librosa.power_to_db(S))
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
p = plt.gca()
p.yaxis.tick_right()
p.yaxis.set_label_position("right")
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
